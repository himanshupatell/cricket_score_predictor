```python
df['over'] = df['ball'].apply(lambda x : str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x : str(x).split(".")[1])
```

```python
df.head()
```

```python

df['ball_bowled'] = (df['over'].fillna(0).astype(int) * 6 + df['ball_no'].fillna(0).astype(int))
```

```python
df.tail()
```

```python
df['balls_left'] = 120 - df['ball_bowled']
```

```python
df.tail()
```

```python
import numpy as np
df['balls_left'] = np.where(df['balls_left'] < 0, 0, df['balls_left'])
```

```python
df.tail()
```

```python
df.info()
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:1 if x!='0' else '0')
```

```python
df
```

```python
df['player_dismissed'] = df['player_dismissed'].astype(int)
```

```python
# Create a new column where dismissal is 1 else 0
df['is_wicket'] = df['player_dismissed'].notnull().astype(int)

# Cumulative sum of dismissals per match
df['wickets_fallen'] = df.groupby('match_id')['is_wicket'].cumsum()
```

```python
df['wicket_left'] = 10 - df['player_dismissed']
```

```python
df.tail()
```

```python
# STEP 1 - Create current_score
df['current_score'] = df.groupby('match_id')['runs'].cumsum()

# STEP 2 - Create ball_bowled
df['ball_bowled'] = (df['over'].astype(int) * 6 + df['ball_no'].astype(int))

# STEP 3 - Now create current_run_rate
df['current_run_rate'] = (df['current_score'] * 6) / df['ball_bowled']
```

```python
df.tail()
```

```python
# 1. Sort by match and ball order
df = df.sort_values(by=['match_id', 'over', 'ball_no'])

# 2. Calculate last_five runs
groups = df.groupby('match_id')
match_ids = df['match_id'].unique()

last_five = []

for id in match_ids:
    match = groups.get_group(id)
    runs_rolling = match['runs'].rolling(window=30, min_periods=1).sum()
    last_five.extend(runs_rolling.tolist())

# 3. Assign back to DataFrame
df['last_five'] = last_five
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
transformer = ColumnTransformer([
    ('transformer', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
], remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1', transformer),
    ('step2', StandardScaler()),
    ('step3', XGBRegressor(n_estimators=1000, learning_rate=0.2, max_depth=12, random_state=1))
])
```

```python
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```
